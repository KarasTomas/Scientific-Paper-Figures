# Example of template usage

Creation of simple line chart with max 3 datasets - shown on Stress-Strain diagram

## Imports

```python
import sys
import os
sys.path.append(os.path.abspath('..'))
from templates import line_chart as lc

notebook_dir = os.path.dirname(os.path.abspath(__file__)) if '__file__' in globals() else os.getcwd()
project_root = os.path.abspath(os.path.join(notebook_dir, '..'))
```

## Data selection

```python
import pandas as pd

data_file = os.path.join(project_root, "data", "stress_strain_data.csv")
output_file = os.path.join(project_root, "figures", "stress_strain_plot.svg")

# Load the data
data = pd.read_csv(data_file)

# Build the datasets dict
datasets = {
    "Steel": {"x": data["Strain"], "y": data["Stress_Steel"], "label": "Steel"},
    "Aluminium": {"x": data["Strain"], "y": data["Stress_Aluminium"], "label": "Aluminium Alloy"},
    "Titanium": {"x": data["Strain"], "y": data["Stress_Titanium"], "label": "Titanium Alloy"},
}
```

## Plot creation

```python
# Generate the plot
lc.create_3line_plot(
    datasets=datasets,
    output_path=output_file,
    x_label="Strain (mm/mm)",
    y_label="Stress (MPa)",
    title="Stress-Strain Diagram for Three Materials",
)
```

```python
import pandas as pd

data_file = os.path.join(project_root, "data", "stress_strain_data.csv")
output_file = os.path.join(project_root, "figures", "stress_strain_plot.svg")

# Load the data
data = pd.read_csv(data_file)

datasets = {
    "Steel": {"x": data["Strain"], "y": data["Stress_Steel"], "label": "Steel"},
    # "Aluminium": {"x": data["Strain"], "y": data["Stress_Aluminium"], "label": "Aluminium Alloy"},
    "Titanium": {"x": data["Strain"], "y": data["Stress_Titanium"], "label": "Titanium Alloy"},
    "Low Alloy Steel": {"x": data["Strain"], "y": data["Stress_low_alloy_steel"], "label": "Low Alloy Steel"},
    "Cast Iron": {"x": data["Strain"], "y": data["Stress_cast_iron"], "label": "Cast Iron"},
    "S235 Steel": {"x": data["Strain"], "y": data["Stress_S235_steel"], "label": "S235 Steel"},
}

# Generate the plot
lc.create_5line_plot(
    datasets=datasets,
    output_path=output_file,
    x_label=r"Strain ($\epsilon$, mm/mm)",  # Example with epsilon for strain
    y_label=r"Stress ($\sigma$, MPa)",     # Sigma for stress
    title="Stress-Strain Diagram for Three Materials",
)
```
